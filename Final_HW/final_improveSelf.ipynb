```r
# load package and data
library(ggplot2)
```

```r
pm25 <- read.csv("pm25_202112.csv")
pm25
```

```r
# new version
theme_set(theme_bw())

# Scatterplot
gg <- ggplot(pm25, aes(x=date, y=Average)) + 
  geom_point(aes(col=date, size=Average)) + 
  labs(subtitle="December, 2021", 
       y="Average", 
       x="date", 
       title="PM2.5 Concentration Trend", 
       caption = "Source: https://github.com/Yvette0828/111-1-econDV/tree/main/Week15_HW") +
  theme(axis.text.x = element_text(angle = 90, vjust=0.5))

plot(gg)
```
